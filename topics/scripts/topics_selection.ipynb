```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 11
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = 'Arial'
```

```python
df_out = pd.read_csv('inertias.csv')

fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(df_out['n_topics'], df_out['inertia'], color = '#C2C5AA')
ax.plot(df_out['n_topics'], df_out['inertia'], '.', color = '#333D29')
ax.set_xlabel('# topics')
ax.set_ylabel('inertia')
ax.set_title('Within-Cluster Sum of Squares')
ax.set_xticks(ticks=[5, 50, 100, 200, 300, 400])

plt.tight_layout()
plt.savefig('../images/topics_elbow.pdf')
```
